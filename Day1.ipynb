```python
#Write a program that keeps accepting numbers from user. If sum of accepted numbers is more than 300, the program execution should stop.

s = 0

while True:
    num = int(input("Enter a number: "))
    s += num
    if s>300:
        print("the sum exceeded",s)
        break
        
```
